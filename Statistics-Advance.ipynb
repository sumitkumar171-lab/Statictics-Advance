```python
'''
Q-1> What is a random variable in probability theory ?
Sol:
A random variable is a function that assigns a numerical value to each possible outcome of a random experiment. It essentially maps outcomes from
a sample space to real numbers.
Types:
1.Discrete Random Variable – Takes on a countable number of values (e.g., the number of heads in 10 coin flips).
2.Continuous Random Variable – Takes on an infinite number of values within a range (e.g., the time it takes for a bus to arrive).
Ex:
1.Rolling a die: The outcome (1, 2, 3, 4, 5, or 6) can be represented by a random variable X.
2.Measuring temperature: The value could be any real number within a range, making it a continuous random variable.
3. using code:
'''
import numpy as np
from scipy.stats import rv_discrete
values = [1, 2, 3, 4, 5, 6]  
probabilities = [1/6] * 6  
die = rv_discrete(name='die',values=(values,probabilities))
samples = die.rvs(size=10)  
print(samples)
print("P(X=3):",die.pmf(3))

```

```python
'''
Q-5>  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF) ?
Sol:
The Cumulative Distribution Function (CDF) and Probability Density Function (PDF) are both related to the probability distribution of a 
random variable, but they describe different aspects of the distribution.
Probability Density Function (PDF):
1.A PDF describes the likelihood or density of a random variable taking a specific value.
2.It is used for continuous random variables.
3.The PDF provides the probability density, not the actual probability.
4.The probability of X falling in a specific range [a,b] is given by the area under the curve between a and b.
5.The PDF provides the probability density, not the actual probability.
6.The probability of X falling in a specific range [a,b] is given by the area under the curve between a and b.
For any specific value x, the probability is zero for continuous distributions, because the probability of a continuous random variable 
taking a single value is infinitesimally small.
Cumulative Distribution Function (CDF):
1.A CDF gives the cumulative probability that a random variable X is less than or equal to a particular value x.
2.It is used for both discrete and continuous random variables.
3.The CDF provides the cumulative probability up to a specific point x.
4.It gives the probability that the random variable will take a value less than or equal to x.
5. For continuous random variables, the CDF is always non-decreasing and ranges from 0 to 1.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu = 0        
sigma = 1     
x = np.linspace(-5, 5, 1000)
pdf_values = norm.pdf(x, loc=mu, scale=sigma)
cdf_values = norm.cdf(x, loc=mu, scale=sigma)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, label="PDF", color="blue")
plt.fill_between(x, pdf_values, alpha=0.2, color="blue")
plt.title("Probability Density Function (PDF)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(x, cdf_values, label="CDF", color="red")
plt.fill_between(x, cdf_values, alpha=0.2, color="red")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
'''
Q-6> What is a discrete uniform distribution ?
Sol:
A Discrete Uniform Distribution is a type of probability distribution where all possible outcomes are equally likely. It is used when a 
random variable takes on a finite number of distinct values, and each of those values has the same probability of occurring.
In other words, for a discrete uniform distribution, every possible outcome in the sample space has an equal chance of happening.
Key:
1.Finite Number of Outcomes
2.Equal Probability for All Outcomes
3.Probability Mass Function (PMF)
'''
import numpy as np
import matplotlib.pyplot as plt
a = 1  # minimum value
b = 6  # maximum value
x = np.arange(a, b + 1)
probabilities = np.ones(len(x)) / len(x)
plt.stem(x, probabilities, basefmt=" ", label="PMF")
plt.xlabel("Outcomes (X)")
plt.ylabel("P(X=x)")
plt.title("Discrete Uniform Distribution (Fair Die)")
plt.xticks(x)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()

```

```python
'''
Q-7> What are the key properties of a Bernoulli distribution ?
Sol:
A Bernoulli distribution is a discrete probability distribution that describes a random experiment with exactly two possible outcomes: 
success (often labeled as 1) and failure (labeled as 0). It is named after the Swiss mathematician Jacob Bernoulli.
Key Properties :
1.Two Possible Outcomes
2.Probability Mass Function (PMF)
3.Mean (Expected Value)
4.Variance
5.Standard Deviation
6.Mode
7.Skewness
8. Entropy
'''
import matplotlib.pyplot as plt
import numpy as np
p = 0.5
outcomes = [0, 1]
probabilities = [1 - p, p]
plt.bar(outcomes, probabilities, width=0.2, color='blue', alpha=0.7)
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title(f"Bernoulli Distribution (p = {p})")
plt.xticks(outcomes)
plt.ylim(0, 1)
plt.show()
```

```python
'''
Q-8> What is the binomial distribution, and how is it used in probability ?
Sol:
The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent
Bernoulli trials (experiments), each with the same probability of success.
It is one of the most commonly used probability distributions in statistics, particularly in scenarios where there are binary outcomes
(success or failure) and the number of trials is fixed.
Characteristics :
1.Fixed Number of Trials (n)
2.Two Possible Outcomes
3.Constant Probability of Success (p)
4.Independence
Ex:
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
n = 10  
p = 0.5  
k = 6    
prob_6_heads = binom.pmf(k, n, p)
print(f"Probability of getting exactly 6 heads: {prob_6_heads:.4f}")
x = np.arange(0, n + 1)
pmf_values = binom.pmf(x, n, p)
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.xlabel("Number of Successes (Heads)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.xticks(x)
plt.show()
```

```python
'''
Q-9> What is the Poisson distribution and where is it applied ?
Sol:
The Poisson distribution is a discrete probability distribution that models the number of events occurring within a fixed interval of 
time or space, given that:
The events happen independently.
The rate at which events occur is constant (i.e., they occur at an average rate λ over time or space).
The number of events in one interval does not affect the number of events in another interval.
The events happen rarely (i.e., the number of events in any given interval is small compared to the number of possible occurrences).
It is named after the French mathematician Siméon-Denis Poisson.
Keys:
1.Number of Events
2.Rate of Occurrence (λ)
3.Probability Mass Function (PMF)
4.Mean and Variance
5.Skewness and Shape
Ex:
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
lambda_ = 3 
k = 4        
prob_4_customers = poisson.pmf(k, lambda_)
print(f"Probability of exactly 4 customers: {prob_4_customers:.4f}")
k_values = np.arange(0, 11)
pmf_values = poisson.pmf(k_values, lambda_)
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.xlabel("Number of Customers (k)")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution PMF (λ={lambda_})")
plt.xticks(k_values)
plt.show()
```

```python
'''
Q-11> What are the characteristics of a normal distribution ?
Sol:
A normal distribution, also known as a Gaussian distribution, has the following key characteristics:
1.Bell-shaped curve – The distribution is symmetric and has a single peak at the mean.
2.Symmetry – The left and right sides of the distribution are mirror images of each other.
3.Mean, Median, and Mode are Equal – All three measures of central tendency are located at the center of the distribution.
4.Defined by Mean (μ) and Standard Deviation (σ) – The shape and spread of the normal distribution depend on these two parameters:
     1.μ (mean) determines the center of the distribution.
     2. σ (standard deviation) controls the spread (wider for larger σ, narrower for smaller σ).
5.68-95-99.7 Rule (Empirical Rule) – In a normal distribution:
    1~68% of data falls within 1σ of the mean.
    2.~95% of data falls within 2σ of the mean.
    3.~99.7% of data falls within 3σ of the mean.
6.Asymptotic Nature – The tails of the distribution extend infinitely in both directions but never touch the horizontal axis.
7.Total Area Under the Curve = 1 – This represents the probability of all possible values.
8.No Skewness or Kurtosis – A normal distribution has a skewness of 0 (perfectly symmetric) and a kurtosis of 3 (mesokurtic).
'''
```

```python
'''
Q-12> What is the standard normal distribution, and why is it important ?
Sol:
The standard normal distribution is a special case of the normal distribution where:
The mean (μ) = 0
The standard deviation (σ) = 1
Mathematically, a normal variable X can be converted to a standard normal variable Z using the Z-score formula:
Z= X−μ/σ
Where,
where:
Z is the standard normal variable.
X is the raw score.
μ is the mean.
σ is the standard deviation.
Keys:
1.Simplifies Probability Calculations:
Probability tables (Z-tables) provide cumulative probabilities for standard normal values, making it easier to compute probabilities 
without complex integration.
2.Allows for Comparisons Across Different Datasets:
By converting any normal distribution to a standard normal distribution using Z-scores, we can compare values from different distributions.
3.Foundation for Inferential Statistics
Many statistical tests, such as hypothesis testing and confidence intervals, rely on the standard normal distribution.
4.Central in the Central Limit Theorem (CLT)
The CLT states that, for a sufficiently large sample size, the sampling distribution of the mean approaches a normal distribution 
(often standard normal after standardization).
'''
```

```python
'''
Q-13> What is the Central Limit Theorem (CLT), and why is it critical in statistics ?
Sol:
The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases,
regardless of the shape of the original population distribution, provided the samples are independent and identically distributed (i.i.d.).
'''
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(5, 2))
ax.axis("off")
formula_text = r"$E(\bar{X}) = \mu$" + "\n\n" + r"$\sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}$"ax.text(0.5, 0.5, formula_text, fontsize=16, ha="center", va="center")
plt.show()

'''
where:
μ is the population mean.
𝜎 is the standard deviation of the sample mean (also called the standard error).
Keys:
1.Enables Inferential Statistics
The CLT allows us to make inferences about a population using sample data, even when the population distribution is unknown or non-normal.
2.Justifies the Use of Normal Distribution in Many Tests
Many statistical methods (e.g., hypothesis testing, confidence intervals) assume normality. The CLT ensures that this assumption holds for sufficiently large samples.
3.Simplifies Probability Calculations
Instead of using complex probability distributions, the normal distribution (with Z-scores) can be used to approximate probabilities of sample means.
4.Works Even for Skewed or Non-Normal Populations
Even if the original population is skewed, bimodal, or heavily tailed, the distribution of sample means will still be approximately normal for large 
n (typically n≥30).
'''
```

```python
'''
Q-15> What is the application of Z statistics in hypothesis testing ?
Sol:
Z-statistics (or Z-tests) are used in hypothesis testing when the population variance is known and the sample size is large (n≥30). It 
is based on the standard normal distribution (Z-distribution).
Steps :
1. Define the Hypotheses
2.Compute the Test Statistic
3.Determine the Critical Value or p-Value
4.Make a Decision
Types:
1.One-Sample Z-Test: Tests whether the mean of a single sample differs from a known population mean.
2.Two-Sample Z-Test: Compares the means of two independent groups.
3.Proportion Z-Test: Tests for differences in population proportions.
Z-test using python:
'''
from scipy.stats import norm
X_bar=129.4   
mu=200     
sigma=980    
n=90        
alpha=0.56  
Z = (X_bar - mu)/(sigma/(n ** 0.5))
p_value = 2*(1 - norm.cdf(abs(Z)))
reject_H0 = p_value<alpha
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Reject Null Hypothesis? {'Yes' if reject_H0 else 'No'}")
```
